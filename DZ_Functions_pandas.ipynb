# Задание №1

```python
import pandas as pd
```

```python
"""Напишите функцию, которая классифицирует фильмы из материалов занятия по следующим правилам:
- оценка 2 и меньше - низкий рейтинг
- оценка 4 и меньше - средний рейтинг
- оценка 4.5 и 5 - высокий рейтинг
Результат классификации запишите в столбец class"""

data = pd.read_csv('ratings.csv')
data_1 = pd.read_csv('movies.csv')
data_2 = data.merge(data_1, how='left', on='movieId')
data_2.head()
```

```python
def classification(grade):
    if grade <= 2:
        class_ = 'Низкий рейтинг'
        
    elif 2 <= grade <= 4:
        class_ = 'Средний рейтинг'
        
    elif grade >= 4:
        class_ = 'Высокий рейтинг'
    
    return  class_
    
data_2['class'] = data_2['rating'].apply(classification)
data_2.head()
```

# Задание №2

```python
"""Необходимо написать гео-классификатор, который каждой строке сможет выставить географическую принадлежность 
определенному региону. Т. е. если поисковый запрос содержит название города региона, то в столбце ‘region’
пишется название этого региона. Если поисковый запрос не содержит названия города, то ставим ‘undefined’."""

keywords = pd.read_csv('keywords.csv')
keywords.head(70)
```

```python
"""Пишем функцию"""
def region (keyword):
    if 'моск'in  keyword or 'тул'in keyword or 'яросл' in keyword:
        region_ = 'Центр'
    elif 'петер'in  keyword or 'псков' in keyword or 'мурм' in keyword:
        region_ = 'Северо-Запад'
    elif 'владив' in  keyword or 'сахал' in keyword or 'хабаровс' in keyword:
        region_ = 'Дальний Восток'
    else:
        region_ = 'undefined'
    return region_

keywords['region'] = keywords['keyword'].apply(region)
keywords.head()
```

```python
"""проверка"""
keywords.groupby('region').count().head()
```

# Задание №3

```python
data = pd.read_csv('ratings.csv')
data_1 = pd.read_csv('movies.csv')
data_2 = data.merge(data_1, how='left', on='movieId')
data_2.head() 
```

```python
"""В переменную years запишите список из всех годов с 1950 по 2010"""
years = []
years = [(lambda x: x )(x) for x in range (1950,2011)]

"""если номер года присутствует в названии фильма, то функция возвращает этот год как год выпуска
если ни один из номеров года списка years не встретился в названии фильма, то возвращается 1900 год"""
def production_year (title):
    for year in years:
        if str(year) in title:
            return str(year)
    return '1900'

"""Запишите год выпуска фильма по алгоритму пункта 2 в новый столбец ‘year’"""

data_2 ['year'] = data_2 ['title'].apply(production_year)
data_2.head()
```

```python
"""Посчитайте средний рейтинг всех фильмов для каждого значения столбца ‘year’ и отсортируйте результат по убыванию рейтинга"""
data_2.groupby('year').agg({'rating': 'mean'}).sort_values('rating', ascending=False).head(50)
```
